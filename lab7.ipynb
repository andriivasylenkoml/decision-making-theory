```python
from graphviz import Digraph

# Створення графа
dot = Digraph(comment='Дворівневе дерево рішень (повна відповідність фото)')
dot.attr(rankdir='LR', size='15,9', nodesep='0.3', ranksep='1.2')

# --- Вузли ---

# Вузол рішення про аудит
dot.node('D1', '1\nРішення про аудит', shape='square')

# Вузол шансу A
dot.node('A', 'A\n2764.72 грн\n(до -80 грн)', shape='circle')

# Вузли рішень після аудиту
dot.node('D2', '2\n2962.60 грн', shape='square')
dot.node('D3', '3\n2154.95 грн', shape='square')

# Вузол рішення без аудиту
dot.node('D4', '4\n2764.41 грн', shape='square')

# Вузли шансів B–G
dot.node('B', 'B', shape='circle')
dot.node('C', 'C', shape='circle')
dot.node('D', 'D', shape='circle')
dot.node('E', 'E', shape='circle')
dot.node('F', 'F', shape='circle')
dot.node('G', 'G', shape='circle')

# Листя (кінцеві вузли з грошовими результатами)
dot.node('B_ret', 'Повернення\n20130 грн', shape='circle')
dot.node('B_def', 'Не повернення\n0 грн', shape='circle')

dot.node('C_ret', 'Безризик. інвест.\n17985 грн', shape='circle')

dot.node('D_ret', 'Повернення\n20130 грн', shape='circle')
dot.node('D_def', 'Не повернення\n0 грн', shape='circle')

dot.node('E_ret', 'Безризик. інвест.\n17985 грн', shape='circle')

dot.node('F_ret', 'Повернення\n20130 грн', shape='circle')
dot.node('F_def', 'Не повернення\n0 грн', shape='circle')

dot.node('G_ret', 'Безризик. інвест.\n17985 грн', shape='circle')

# --- Зв'язки (гілки) ---

# Від рішення про аудит
dot.edge('D1', 'A', label='Проводити аудит', xlabel='(-80 грн)', style='dashed')
dot.edge('D1', 'D4', label='Без аудиту', style='dashed')

# Від A
dot.edge('A', 'D2', label='Рекомендовано\n(0.755)')
dot.edge('A', 'D3', label='Не рекомендовано\n(0.245)')

# Від D2
dot.edge('D2', 'B', label='Давати позику', xlabel='(-16500)', style='dashed')
dot.edge('D2', 'C', label='// Не давати позику', xlabel='(-16500)', style='dashed')

# Від D3
dot.edge('D3', 'D', label='Давати позику', xlabel='(-16500)', style='dashed')
dot.edge('D3', 'E', label='// Не давати позику', xlabel='(-16500)', style='dashed')

# Від D4
dot.edge('D4', 'F', label='Давати позику', xlabel='(-16500)', style='dashed')
dot.edge('D4', 'G', label='// Не давати позику', xlabel='(-16500)', style='dashed')

# --- Кінцеві гілки шансів ---

# B
dot.edge('B', 'B_ret', label='Повернуть (0.9669)')
dot.edge('B', 'B_def', label='Не повернуть (0.0331)')

# C
dot.edge('C', 'C_ret', label='Інвестувати (1.0)', style='dashed')

# D
dot.edge('D', 'D_ret', label='Повернуть (0.9265)')
dot.edge('D', 'D_def', label='Не повернуть (0.0735)')

# E
dot.edge('E', 'E_ret', label='Інвестувати (1.0)', style='dashed')

# F
dot.edge('F', 'F_ret', label='Повернуть (0.957)')
dot.edge('F', 'F_def', label='Не повернуть (0.043)')

# G
dot.edge('G', 'G_ret', label='Інвестувати (1.0)', style='dashed')

# --- Візуалізація ---
dot.render('full_decision_tree_lab7', view=False, format='png')

print("Граф 'full_decision_tree_lab7.png' успішно створено.")
```

```python
from graphviz import Digraph

# Створення графа
dot = Digraph(comment='Фінальне дерево рішень для звіту')
dot.attr(rankdir='LR', size='10,6', ranksep='1.2')

# --- Вузли ---

# Вузол прийняття рішення
dot.node('D1', '[1] Прийняття рішення\nВидача позики', shape='square', xlabel='1617')

# Кінцеві вузли (листя) з результатами
dot.node('A_res1', '20130 грн', shape='circle')
dot.node('A_res2', '0 грн', shape='circle')
dot.node('B_res1', '17985 грн', shape='circle')

# --- Група для вузла А (число + коло) ---
with dot.subgraph(name='cluster_A') as c:
    c.attr(style='invis')
    c.node('A_label', '1617', shape='plaintext')
    c.node('A', 'A', shape='circle')
    c.edge('A_label', 'A', style='invis')

# --- Група для вузла B (число + коло) ---
with dot.subgraph(name='cluster_B') as c:
    c.attr(style='invis')
    c.node('B_label', '1485', shape='plaintext')
    c.node('B', 'B', shape='circle')
    c.edge('B_label', 'B', style='invis')

# --- Зв'язки (гілки) ---

# Від рішення
dot.edge('D1', 'A',
         label='Давати позику під 22%',
         xlabel='(-16500 грн)',
         style='dashed,bold')

dot.edge('D1', 'B',
         label='// Інвестувати під 9%',
         xlabel='❌',
         style='dashed',
         color='red')

# Від A
dot.edge('A', 'A_res1', label='Повернуть (0.9)', style='bold')
dot.edge('A', 'A_res2', label='Не повернуть (0.1)')

# Від B
dot.edge('B', 'B_res1', label='Інвестувати (1.0)', style='dashed')

# --- Візуалізація ---
dot.render('final_decision_tree_lab7', view=False, format='png')

print("Граф 'final_decision_tree_lab7.png' успішно створено.")
```
